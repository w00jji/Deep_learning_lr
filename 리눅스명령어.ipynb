## 반드시 왼쪽 작업디렉토리를 모니터링하면서 진행

```python
# !와 % 차이점

느낌표(!)는 조회, 생성 같이 한번만 실행할때 쓰는 magic line 언어이다. 단, 경로 이동은 적용되지 않는다.
퍼센테이지(%)는 실제 경로 이동을 반영할때 주로 쓰이며, 모든 쉘 언어 사용 가능
```

```python
%cd ..
```

```python
# 현재 작업 경로 출력
!pwd
```

```python
%cd /content
```

```python
# 디렉토리 만들기
!mkdir test
```

```python
# 현재 작업 경로에 있는 모든 파일 확인
!ls
```

```python
# 폴더 이동
%cd test
```

```python
# 파일 한번 만들어보기(이론 설명 후 실습)
!mkdir IMBK
```

```python
!pwd
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# 관리자랑 그룹 소속들은 못건들고 나만 읽고쓰고 실행 가능하게 할거다.
# rwx <- 421
# drwxr-xr-x <- 755
!chmod 007 osh.txt
```

```python
# 디렉토리 만들기
!mkdir a
!mkdir b
!mkdir c
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# a 디렉토리의 권한 수정하기
!chmod 700 a
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# a 파일을 상위 디렉토리로 옮기기

# 규칙
#mv 어떤 폴터? 어디로?
!mv ./a ../
```

```python
# c 파일을 상위 디렉토리로 복사하기

#cp 할 때 폴터 안쪽에 파일이 있어도 복사 가능하게 해주는
# -r 옵션.
!cp -r ./c ../
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# c 파일을 b 디렉토리로 복사하기
!cp -r ./c ./b
```

```python
# b 디렉토리 안에 있는 정보 보기
!ls -al ./b
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```

```python
# 파일 삭제하기, 오류가 날 것 이다.'

# 왜냐 b폴더안에 c폴더가 있기 때문이다.
!rm b
```

```python
# 폴더 또는 그 폴더 안에 있는 파일도 삭제할 때 -r 옵션을 사
!rm -r b
```

```python
# 모든 파일의 모든 정보 보기
!ls -al
```
